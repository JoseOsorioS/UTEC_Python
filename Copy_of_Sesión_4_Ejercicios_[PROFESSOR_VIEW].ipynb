# Ejercicio 1. Gráficos de Funciones

```python
# Ejercicio 1.1
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = x ** 2

plt.plot(x, y, label='y = x^2')
plt.title('Gráfico de la función y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Ejercicio 1.2
x = np.linspace(0, 2 * np.pi, 400)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='y = sin(x)')
plt.plot(x, y_cos, label='y = cos(x)')
plt.title('Gráfico de las funciones y = sin(x) y y = cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Ejercicio 1.3
x = np.linspace(-2, 2, 400)
y = np.exp(x)

plt.plot(x, y, label='y = e^x')
plt.title('Gráfico de la función y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from IPython.display import display
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET

df = pd.read_excel(PATH + "American Ninja Warrior Obstacle History.xlsx")


df.head()
```

## Histograma del Orden de los Obstáculos en American Ninja Warrior

```python
import matplotlib.pyplot as plt

# Suponiendo que tienes un dataframe llamado df
plt.hist(df['Obstacle_Order'], bins=10, edgecolor='white')
plt.xlabel('Orden de los Obstáculos')
plt.ylabel('Frecuencia')
plt.title('Histograma del Orden de los Obstáculos en American Ninja Warrior')
plt.show()
```

## Gráfico de Líneas

```python
obstaculos_por_temporada = df.groupby('Season')['Obstacle_Order'].mean()
temporadas = obstaculos_por_temporada.index
plt.plot(temporadas, obstaculos_por_temporada, marker='o')
plt.xlabel('Temporada')
plt.ylabel('Orden Promedio de Obstáculos')
plt.title('Evolución del Orden Promedio de Obstáculos por Temporada')
plt.xticks(temporadas)
plt.grid(True)
plt.show()
```

# Subplots

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Histograma
ax1.hist(df['Obstacle_Order'], bins=10, edgecolor='black')
ax1.set_xlabel('Orden de los Obstáculos')
ax1.set_ylabel('Frecuencia')
ax1.set_title('Histograma del Orden de los Obstáculos')

# Gráfico de líneas
obstaculos_por_temporada = df.groupby('Season')['Obstacle_Order'].mean()
temporadas = obstaculos_por_temporada.index
ax2.plot(temporadas, obstaculos_por_temporada, marker='o')
ax2.set_xlabel('Temporada')
ax2.set_ylabel('Orden Promedio de Obstáculos')
ax2.set_title('Evolución del Orden Promedio de Obstáculos por Temporada')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

## Diagrama de Caja
Para un diagrama de caja del orden de los obstáculos:

```python
plt.boxplot(df['Obstacle_Order'])
plt.xlabel('Orden de los Obstáculos')
plt.ylabel('Valor')
plt.title('Diagrama de Caja del Orden de los Obstáculos')
plt.show()
```

## Gráfico de Pie

```python
conteo_rondas = df['Round_or_Stage'].value_counts()
plt.pie(conteo_rondas, labels=conteo_rondas.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribución de Rondas o Etapas en American Ninja Warrior')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/germaingarcia/diplomadoFiles/main/conjunto_datos%20(2).csv')
df.head()
```

```python
df.corr()
```

```python
df_frienzone = df[df['FRIENZONEADO']==1]
df_NO_frienzone = df[df['FRIENZONEADO']==0]
```

```python
df.columns
```

```python
plt.figure(figsize=(8, 5))

# Plotting the boxplot
lista = ["Grado_Simpatico","Regalos_Realizados","Te_habla_de_sus_citas","Tiempo_dedicado","Amigos_presentados","Veces_que_quedais_a_solas","Consejos_dados"]
df_frienzone[lista].boxplot()

plt.title('Box Plot para Friendzoneado')
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
df_NO_frienzone[lista].boxplot()
plt.title('Box Plot para NO frienzoneado')
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
